```
import pandas as pd

df = pd.read_csv('/Users/humberto/Desktop/MODULO7_PROJETOFINAL_BASE_SUPERMERCADO.csv')
print(df.head(10))

#Cálculo para saber quais Marcas tem mais produtos

df.groupby('Marca')['title'].count().reset_index().sort_values(by='title', ascending=False)
print(df.groupby('Marca')['title'].count().head(10))

#Cálculo para saber quais Categorias tem mais produtos

df.groupby('Categoria')['title'].count().reset_index().sort_values(by='title', ascending=False)
print(df.groupby('Categoria')['title'].count())

##Para os cálculos estatísticos vou seguir com a coluna Categoria visto que ela possui mais valores numéricos atrelados á ela.

#Porcentagem de produtos por categoria

# Dessa forma abaixo trazemos a porcentagem de cada esporte
contagem_por_categoria = df['Categoria'].value_counts(normalize=True).reset_index()
print(contagem_por_categoria)

# 1 - Média

soma_precos_totais = df['Preco_Normal'].sum()
quantidade_de_registros = len(df['Preco_Normal'])
media_preco_normal = soma_precos_totais / quantidade_de_registros

print("Média de preços totais:", media_preco_normal)

#Formula mais rápida e simples d caulo da média
media = df['Preco_Normal'].mean()
print(media)

#Média de preços por categoria
media_preco_categoria = df.groupby('Categoria')['Preco_Normal'].mean().reset_index().sort_values(by='Preco_Normal', ascending=False)
print(media_preco_categoria)

# 2 - Mediana

mediana = df['Preco_Normal'].median()
print('A mediana dos preços é:', mediana)

##A mediana esta bem abaixo da média, com isso é possível concluir que a possuimos uma distribuição bem disforme pois temos valores com grandes disparidades puxando a média acima da mediana.

#Mediana de preços por categoria
mediana_preco_categoria = df.groupby('Categoria')['Preco_Normal'].median().reset_index().sort_values(by='Preco_Normal', ascending=False)
print(mediana_preco_categoria)

# 3 - Moda

moda = df['Preco_Normal'].mode()
print('A moda dos preços é:', moda)

##O valor de preço que mais se repete na teblo é o 0, evento que ocorre muto provavelmente a preços que não foram imputados ao produto ou produtos que, apesar de estarem na tabela, não estão á venda.

# 4 - Desvio Padrão

desvio_padrao = df['Preco_Normal'].std()
print('O desvio padrão dos preço é:', desvio_padrao)

#Desvio Padrão por categoria de produto
desvio_padrao_por_categoria = df.groupby('Categoria')['Preco_Normal'].std().reset_index()
print( desvio_padrao_por_categoria)

##Aqui conseguimos concluir que o desvio padrão esta bem acima da média. Analisando o DataFrame do Desvio Padrão por categoria isso se dá devido ao preço dos produtos lactéos, pois alêm de estarem bem acima da média ele é a categoria que mais tem produtos disponíveis, o que ê nos fornecido no DataFrame de categoria por quantidade de produtos.

# Final: método Describe

metodo_describe = df['Preco_Normal'].describe()
print(metodo_describe)
```
