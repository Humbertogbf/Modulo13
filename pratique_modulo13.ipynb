```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px

df = pd.read_csv('/Users/humberto/Desktop/MODULO7_PROJETOFINAL_BASE_SUPERMERCADO.csv')
print(df.head(10))

#Cálculo para saber quais Marcas tem mais produtos

df.groupby('Marca')['title'].count().reset_index().sort_values(by='title', ascending=False)
print(df.groupby('Marca')['title'].count().head(10))

#Cálculo para saber quais Categorias tem mais produtos

df.groupby('Categoria')['title'].count().reset_index().sort_values(by='title', ascending=False)
print(df.groupby('Categoria')['title'].count())

##Para os cálculos estatísticos vou seguir com a coluna Categoria visto que ela possui mais valores numéricos atrelados á ela.

#Porcentagem de produtos por categoria

# Dessa forma abaixo trazemos a porcentagem de cada esporte
contagem_por_categoria = df['Categoria'].value_counts(normalize=True).reset_index()
print(contagem_por_categoria)

# 1 - Média

soma_precos_totais = df['Preco_Normal'].sum()
quantidade_de_registros = len(df['Preco_Normal'])
media_preco_normal = soma_precos_totais / quantidade_de_registros

print("Média de preços totais:", media_preco_normal)

#Formula mais rápida e simples d caulo da média
media = df['Preco_Normal'].mean()
print(media)

#Média de preços por categoria
media_preco_categoria = df.groupby('Categoria')['Preco_Normal'].mean().reset_index().sort_values(by='Preco_Normal', ascending=False)
print(media_preco_categoria)

##Com um valor de média aproximado de 2071 conseguimos identificar pelo DataFrame quais produtos estão abaixo ou acima dessa média, em outras palavras, os mais caros e os mais baratos. Dando uma enfâse nas comidas preparadas que esta quase 50% mais cara que a média dos produtos e os instantâneos e sopas que esta um pouco a mais de 1/3 da mésia, sendo a categoria mais em conta.
# 2 - Mediana

mediana = df['Preco_Normal'].median()
print('A mediana dos preços é:', mediana)

#Mediana de preços por categoria
mediana_preco_categoria = df.groupby('Categoria')['Preco_Normal'].median().reset_index().sort_values(by='Preco_Normal', ascending=False)
print(mediana_preco_categoria)

##A mediana esta bem abaixo da média, com isso é possível concluir que a possuimos uma distribuição bem disforme, pois temos valores com grandes disparidades puxando a média acima da mediana. Aqui temos novamente as comidas preparadas e instantâneos e sopas que, respectivamente, estão com 2.6x e 0.34x o valor da mediana. As outras demais categorias se encontram com valores bem próxima da mediana

# 3 - Moda

moda = df['Preco_Normal'].mode()
print('A moda dos preços é:', moda)

##O valor de preço que mais se repete na planilha é o 0, evento que ocorre muto provavelmente a preços que não foram imputados ao produto ou produtos que, apesar de estarem na tabela, não estão á venda.

# 4 - Desvio Padrão

desvio_padrao = df['Preco_Normal'].std()
print('O desvio padrão dos preço é:', desvio_padrao)

#Desvio Padrão por categoria de produto
desvio_padrao_por_categoria = df.groupby('Categoria')['Preco_Normal'].std().reset_index()
print( desvio_padrao_por_categoria)

##Aqui conseguimos concluir que o desvio padrão esta bem acima da média. Analisando o DataFrame do Desvio Padrão por categoria isso se dá devido ao preço dos produtos lactéos, pois alêm de estarem bem acima da média ele é a categoria que mais tem produtos disponíveis, o que ê nos fornecido no DataFrame de categoria por quantidade de produtos.

# 5 - Método Describe

metodo_describe = df['Preco_Normal'].describe()
print(metodo_describe)

# 6 - Boxplot da média e mediana

Q1 = df['Preco_Normal'].quantile(0.25)
Q3 = df['Preco_Normal'].quantile(0.75)
IQR = Q3 - Q1
limite_inferior = Q1 - 1.5 * IQR
limite_superior = Q3 + 1.5 * IQR
print('O IQR, o limite inferior e o superior são respectivamente;', IQR,',', limite_inferior,'e', limite_superior)

outliers = df[(df['Preco_Normal'] < limite_inferior) | (df['Preco_Normal'] > limite_superior)]
numero_de_outliers = outliers.shape[0]
print("Número de outliers:", numero_de_outliers)

plt.figure(figsize=(8, 6))
plt.boxplot(df['Preco_Normal'])
plt.title('Boxplot do Preço Normal', fontsize=16)
plt.ylabel('Valores', fontsize=12)
plt.xlabel('Preço', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()

menor_outlier = outliers['Preco_Normal'].min()
maior_outlier = outliers['Preco_Normal'].max()

print("Menor outlier:", menor_outlier)
print("Maior outlier:", maior_outlier)

##Com base nos dados describe e as informações do Boxplot conseguimos entender que o limite inferior é 0, devido á grande quantidade de produtos com esse valor e que seu limite superior é de 7122.5.
## Outra conclusão que conseguimos chegar também é que a panilha fornecida contém preços com um grande volume de dispersão no sentido de produtos mais caro, já que existem muitos valores acima do limite superior(ouliers), 53 para ser mais exato, e que vão de 7199 até 19788.
##O que puxa esses valores acima do limite superior é a categoria de Comidas Preparadas já que é a única entre as categorias com valores de média e mediana acima do limite superior

# 7 - Desconto por Categoria
media_desconto = df['Desconto'].mean()
print('A media de descontos é:', media_desconto)

media_desconto_categoria = df.groupby('Categoria')['Desconto'].mean().reset_index().sort_values(by='Desconto', ascending=False)
print(media_desconto_categoria)

desconto_por_categoria = df.groupby('Categoria')['Desconto'].mean()
plt.figure(figsize=(10,6))
plt.bar(desconto_por_categoria.index, desconto_por_categoria, color='blue')
plt.title('Desconto por Categoria', fontsize=16)
plt.xlabel('Categorias', fontsize=8)
plt.ylabel('Valores de Desconto', fontsize=12)
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

desconto_categoria = df.groupby('Categoria')['Desconto'].mean().reset_index()
fig = px.bar(desconto_categoria, x='Desconto', y='Categoria', orientation='h',title='Desconto por Categoria',labels={'Desconto': 'Desconto Médio', 'Categoria': 'Categoria'})
fig.show()
##Analisando o Dataframe e o gráfico entende-se que as categorias frutas, instantâneos e verduras não tem nenhum poduto com desconto
##A maior média de desconto em números totais se da para a categoria de congelados com cerca de 7,3% da média de preço normal
##Já comparando com o preço original, a categoria com maior desconto médio é a de beleza e auto cuidado, chegando a 71% do preço normal.

# 8 - Desconto por marca
media_desconto_marca = df.groupby('Marca')['Desconto'].mean().reset_index().sort_values(by='Desconto', ascending=False)
print(media_desconto_marca)

desconto_por_marca = df.groupby('Marca')['Desconto'].mean()
plt.figure(figsize=(10,6))
plt.bar(desconto_por_marca.index, desconto_por_marca, color='blue')
plt.title('Desconto por Marca', fontsize=16)
plt.xlabel('Marca', fontsize=8)
plt.ylabel('Valores de Desconto', fontsize=12)
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

desconto_marca = df.groupby('Marca')['Desconto'].mean().reset_index()
fig = px.bar(desconto_marca, x='Desconto', y='Marca', orientation='h',title='Desconto por Categoria',labels={'Desconto': 'Desconto Médio', 'Marca': 'Marca'})
fig.show()

##Como pudemos analisar, por termos muitas marcas no estoque o gráfico simples em barra fica muito complicado analisar devido aos valores se sobreporem
##Usa-se então um gráfico interativo no qual consegue-se ver com precisão qual marca tem sua média de desconto
##Nesse contexto destaca-se a marca Magnum coma a que possui a maior média d desconto, sendo de 955
```
